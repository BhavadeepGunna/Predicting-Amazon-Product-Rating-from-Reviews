```python
import pandas as pd 
import selenium 
import numpy as np
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException , StaleElementReferenceException
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.amazon.in/')
```

```python
search_bar=driver.find_element(By.ID,'twotabsearchtextbox')
search_bar.send_keys('head phones')
driver.find_element(By.ID,'nav-search-submit-button').click()

```

```python
product_URL=[]
ratings=[]
review=[]
for i in range(1,6):  
    URL=driver.find_elements(By.XPATH,"//div[@class='a-section']/div/div[2]/div/div/div[1]/h2/a")
    for j in URL:
        product_URL.append(j.get_attribute('href'))
    try:
        driver.execute_script("window.scrollBy(0,300)","")
        next_button=driver.find_element(By.XPATH,"//a[text()='Next']")
        next_button.click()
        time.sleep(2)
    except NoSuchElementException:
        print('not happened')
        pass


    


    
```

```python
len(product_URL)
```

```python
unique_value=set(product_URL)
len(unique_value)
```

```python
ratings.clear()
review.clear()
```

```python
count=0
c1=0
c2=0
c3=0
for i in product_URL:
    driver.get(i)
    try:
        driver.find_element(By.XPATH,"//a[1][@id='acrCustomerReviewLink']").click()
    except NoSuchElementException:
        c2+=1
        pass
    try:
        driver.find_element(By.XPATH,"//a[@class='a-link-emphasis a-text-bold']").click()
    except NoSuchElementException:
        c3+=1
        pass

    start_page=1
    end_page=50
    for page in range(start_page,end_page+1):

        try:
            Reviews=driver.find_elements(By.XPATH,"//div[@class='a-row a-spacing-small review-data']/span/span")
            Ratings=driver.find_elements(By.XPATH,"//div[@class='a-row']/a/i/span[@class='a-icon-alt']")
            if(len(Reviews)!=len(Ratings)):
                try:
                    driver.find_element(By.XPATH,"//a[text()='Next page']").click()
                    time.sleep(2)
                    pass
                except NoSuchElementException:
                    print("notclicked")
                    c1+=1
                    break  
            else:
                for r in Reviews:
                    review.append(r.text.replace('\n',''))
                for i in Ratings:
                    rating=i.get_attribute("innerHTML")
                    ratings.append(rating)
                print(len(ratings),len(review))
            time.sleep(2)
        except NoSuchElementException:
            review.append("no Review")
            ratings.append('No rating')
            count+=1
            pass
            
       
        try:
            driver.find_element(By.XPATH,"//a[text()='Next page']").click()
            time.sleep(2)
        except NoSuchElementException:
            print("notclicked")
            c1+=1
            break      

             
 
```

```python
print(len(review),len(ratings))
```

```python
headphones=pd.DataFrame({'reviews':review,'rating':ratings})
```
